```python
import opendatasets as od
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import os
%matplotlib inline

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 150)
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
od.download('https://www.kaggle.com/jsphyg/weather-dataset-rattle-package')
```

```python
os.listdir('weather-dataset-rattle-package')
```

```python
raw_df = pd.read_csv('weather-dataset-rattle-package/weatherAUS.csv')
```

```python
raw_df
```

```python
raw_df.info()
```

```python
raw_df.dropna(subset=['RainTomorrow'], inplace = True)
```

```python
plt.title('No of rows per year')
sns.countplot(x = pd.to_datetime(raw_df.Date).dt.year)
```

```python
year = pd.to_datetime(raw_df.Date).dt.year
train_df = raw_df[year<2015]
val_df = raw_df[year == 2015]
test_df = raw_df[year>2015]
```

```python
input_cols = list(train_df.columns)[1:-1]
target_col = 'RainTomorrow'
```

```python
train_inputs = train_df[input_cols].copy()
train_outputs = train_df[target_col].copy()
```

```python
train_inputs
```

```python
val_inputs = val_df[input_cols].copy()
val_outputs = val_df[target_col].copy()
```

```python
test_inputs = test_df[input_cols].copy()
test_outputs = test_df[target_col].copy()
```

```python
numeric_cols = np.array(['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm'])
# numeric_cols = pd.DataFrame(numeric_cols)
categorical_cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday']
```

```python
train_inputs.columns
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean').fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
test_inputs[numeric_cols].isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(raw_df[numeric_cols])
```

```python
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
val_inputs.describe().loc[['min','max']]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output = False, handle_unknown = 'ignore').fit(raw_df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
X_train_cols = []
X_train_cols.extend(numeric_cols)
X_train_cols.extend(encoded_cols)

X_train = train_inputs[X_train_cols]
```

```python
X_val_cols = []
X_val_cols.extend(numeric_cols)
X_val_cols.extend(encoded_cols)

X_val = val_inputs[X_val_cols]
```

```python
X_test_cols = []
X_test_cols.extend(numeric_cols)
X_test_cols.extend(encoded_cols)

X_test = test_inputs[X_val_cols]
```

```python
X_test
```

## Training and visualizing decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state = 42)
```

```python
model.fit(X_train, train_outputs)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
train_preds = model.predict(X_train)
```

```python
train_preds
```

```python
train_probs = model.predict(X_train)
```

```python
accuracy_score(train_outputs, train_preds)
```

```python
model.score(X_val, val_outputs)
```

```python
from sklearn.tree import plot_tree, export_text
```

```python
plt.figure(figsize = (80,20))
plot_tree(model, feature_names = X_train.columns, max_depth = 2, filled = True)
```

## Hyperparamters tuning ![EJCrSZw.png.crdownload](EJCrSZw.png)

gini function is the loss function that is used by the decision tree to decide which column should be used for splitting the data, and at which point the column should be split. A lower gini index indicates a better split. A perfect split has a gini index of 0.

```python
model.feature_importances_
```

```python
importance_df = pd.DataFrame({
    'feature' : X_train.columns,
    'importance' : model.feature_importances_
}).sort_values('importance', ascending = False)
```

```python
importance_df.head(10)
```

We can change the hyperparameters and see how does it affect our prediction and the accuracy score we can achieved. But with more and more nodes we will face the issue of overfitting as most of the data will lie in one of the node criteria. So we will get training accuracy of more tha 95% so to make sure we perform good on both training as well as val and test we can tune the hyperparameters to get a desireable accuracy

## Random Forest

A random forest works by averaging the results of several decision trees

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_jobs = -1, random_state = 42)
```

```python
model.fit(X_train, train_outputs)
```

```python
model.score(X_train, train_outputs)
```

```python
model.score(X_val,val_outputs)
```

```python
model.score(X_test, test_outputs)
```

```python
importance_df = pd.DataFrame({
    'feature' : X_train.columns,
    'importance' : model.feature_importances_
}).sort_values('importance',ascending = False)
```
